# General observations <a id='general_observations'></a>
---

In general most of the interactions happen with hydrophobic residues, and some charged and polar as well. In the figure below I superimposed the structure of dn47 for all the complexes (minus 2 that changed in conformation slightly). Then I show as sticks the residues that are involved in interactions. The residues are colored according to a YRB scheme described by [Hagemans et al. (2015)](https://www.frontiersin.org/articles/10.3389/fmolb.2015.00056/full). It basically consists of:

- Carbon atoms not bound to nitrogen or oxygen atoms are colored <span style='background-color:yellow'>yellow</span>
- Oxygens carrying the negative charges in glutamate and aspartate are colored <span style='background-color:red'>red</span>
- Nitrogens carrying the positive charges in lysine and arginine <span style='background-color:blue;color:white'>blue</span>
- All remaining atoms are colored <span style='background-color:white;color:black'>white</span>

See [this figure](https://www.frontiersin.org/files/Articles/164719/fmolb-02-00056-HTML/image_m/fmolb-02-00056-g001.jpg) for the specific colors for each amino acid.

In the figures below it is apparent that the residues most commonly involved in interactions are <span style='background-color:yellow'>hydrophobic</span>
