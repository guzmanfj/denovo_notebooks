```python
import pandas as pd
```

```python
highest_scores = pd.read_pickle('./results/notebook06/highest_scores_all.pkl')
```

```python
highest_scores.shape
```

```python
# Sort by pdockq and reset index
highest_scores = (highest_scores.sort_values(by='pdockq', ascending=False)
                                .reset_index().rename(columns={"index": "complex"}))

# Select only complexes with pdockq >= 0.7
top_complexes = highest_scores[highest_scores['pdockq'] >= 0.7]
```

These are the complexes with both **ipTM >= 0.75** and  **pDockQ >= 0.7**:

# General observations
---

dn18 has a mixture of polar and non-polar residues. It is very flexible in the way in which it is predicted to interact. AlphaFold frequently inserts it into existing folds to extend beta-sheet domains, or to bind to hydrophobic patches in the surfaces of proteins.
